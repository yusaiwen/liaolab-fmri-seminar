```python
def save_gii(data, path) -> None:
    if type(data) is not tuple:
        raise Exception("Should give a tuple for both hemispheres!")
    lh = data[0]
    rh = data[1]

    path_lh = path.replace(".gii", "_hemi-L.gii")
    path_rh = path.replace(".gii", "_hemi-R.gii")
    save_gii_LR(lh, "L", path_lh)
    save_gii_LR(rh, "R", path_rh)

def save_gii_LR(data, hemi, path, label_ref=None, label=None) -> None:
        """ 
        Save as a gii data/label file
        
        
        Parameters
        ----------
        data: np.array whose dimension has to be one
        hemi: L| R
        """
        
        import nibabel as nib
        structure_dic = dict(L="CortexLeft",
                             R="CortexRight")
        structure = structure_dic.get(hemi)
        gii = nib.gifti.gifti.GiftiImage()
        data = nib.gifti.gifti.GiftiDataArray(np.array(data, dtype=np.float32))
        gii.add_gifti_data_array(data)
        gii.meta = nib.gifti.gifti.GiftiMetaData(dict(AnatomicalStructurePrimary=structure))

        if label_ref is not None:
            gii.labeltable = nib.load(label_ref).labeltable
        elif label is not None:
            gii.labeltable = label

        gii.to_filename(str(path))
```

```python
def get_model_indexes(cii, structures):

    # a complete representation of the header
    axes = [cii.header.get_axis(i) for i in range(cii.ndim)]
    
    # brain model axis
    time_axis, brain_model_axis = axes
    
    # select the structure you want
    structure_names = structures # list of structure names
    # check if the structure is legal
    for structure in structure_names:
        if structure not in np.unique(brain_model_axis.name):
            raise Exception("structure is not in the cifti file!")

    
    indeces = (brain_model_axis.name == structure_names[0]) + (brain_model_axis.name == structure_names[1])

    return indeces
            
    # # Iter_structures takes concatenated structures, and returns on at a time
    # brain_models = [bm for bm in brain_model_axis.iter_structures()
    #                if bm[0] in structure_names]


def get_model_value(cii_f, structures=None):
    import nibabel as nib
    from pathlib2 import Path

    if structures is None:
        structures = ["CIFTI_STRUCTURE_CORTEX_LEFT", "CIFTI_STRUCTURE_CORTEX_RIGHT"]
    cii = Path(cii_f)
    if cii.exists() is not True:
        raise Exception("Cifti file doesn't exis!")
    cii = nib.load(cii)
    indeces = get_model_indexes(cii, structures)
    
    return cii.get_fdata()[:, indeces]

# get_axis() returns a BrainModelAxis class which is iterable
# BrainModelAxis class following parameters:
## name
## voxel
## vertex
## affine
## volume_shape
## nvertices
# details of BrainModelAxis class can be seen in https://nipy.org/nibabel/reference/nibabel.cifti2.html#nibabel.cifti2.cifti2_axes.Axis
```

```python
def get_atlas(atlas):
    import nibabel as nib
    nib.imageglobals.logger.level = 40
    
    if type(atlas) in [tuple, list, np.array]:
        atlas = np.array(atlas)
        if atlas.shape[0] == 1:
            atlas = atlas[0]
        elif atlas.shape[0] == 2:
            atlas1 = nib.load(atlas[0]).get_fdata().flatten()
            # notice that the two hemispheres have same indeces for the same areas
            # Take, V1, left-V1 and right-V1 are labelled as 1
            atlas2 = nib.load(atlas[1]).get_fdata().flatten()
            atlas2[atlas2>=1] = atlas2[atlas2>=1] + atlas1.max()
            atlas = np.concatenate([atlas1, atlas2], axis=0)
        else:
            raise Exception("Atlas file outnumbers!!")
    elif type(atlas) == str:
        atlas = nib.load(atlas).get_fdata().flatten()
    else:
        raise Exception("Atlas unknown type!")
    return atlas

def avg_within_roi(tseries, atlas, roi):
    return tseries[:, atlas == (roi+1)].mean(axis=1).tolist()

def cal_fc(tseries, atlas):
    # get atlas
    atlas = get_atlas(atlas)

    # average timeseries within timeseries
    roi_t = []
    for roi in np.arange(atlas.max()):
        roi_t.append(avg_within_roi(tseries, atlas, roi))
    roi_t = np.array(roi_t)
    fc = np.corrcoef(roi_t)
    return fc
```

Read the glasser atlas offered by the HCP.

```python
glasser = ("../../../atlas/glasser2016/Q1-Q6_RelatedParcellation210.L.CorticalAreas_dil_Colors.32k_fs_LR.dlabel.nii",
           "../../../atlas/glasser2016/Q1-Q6_RelatedParcellation210.R.CorticalAreas_dil_Colors.32k_fs_LR.dlabel.nii")
```

Construct the FC for each subject. Each subject's timeseries of ten sessions were concatenated linearly.

```python
fc_mats = np.zeros((10, 360, 360)) # subject * vertexes * vertexes
                   
for sub in np.arange(1, 11):
    print(f"sub-{sub}")
    for ses in np.arange(1,11):
        print(f"\t sessoin-{ses}")
        tseries = "sub-%(sub)02d/sub-MSC%(sub)02d_ses-func%(ses)02d_task-rest_bold_32k_fsLR.dtseries.nii" % locals()
        if ses == 1:
            t = get_model_value(tseries)
        else:
            t2 = get_model_value(tseries)
            t  = np.concatenate([t,t2], axis=0)
    fc_mats[sub-1, :, :] = cal_fc(t, glasser)
    del t


sio.savemat("MSC-fc_glasser2016.mat", {"sfc": fc_mats})
```

# PCA

The PCA work is calculaed in R.

In general, the networks were first flatten to one-dimension vector and concatinated by subjects. Then the concatinated matrix is in the dimension of $129600 \times 9$, where 12960 is the square of 360 and 9 is the number of subject(a subject has low quality images). A PCA algorithm is employed on the dimension of subject to extract nine new "subjects", who are linearly independently from each other and are linear combination of nine original subjects.

⬆️ The scree plot of the PCA. Dimensions refer to the extracted new "subjects" and percentage of explained variance refers the variance explained by such dimension. As we can see, the first dimension account for up to 64.3% variance that takes the lead. And this dimension may be the backbone of FC profile across subjects.
